### Python Project: EDA & Data Visualization – AirBnB Listings 2024 (New York)

**1. Importing all necessary libraries**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

**2. Load the datasets**

```python
data = pd.read_csv("datasets.csv",encoding_errors="ignore")
```

**4. Data cleaning**

```python
# dropping all the missing values
data.isnull().sum()
data.dropna(inplace=True) #dropped all missing vaues
data.isnull().sum() # again checking missing values
```

```python
#checking dupilicates
data.duplicated().sum() # shows no.of duplicates
data[data.duplicated()] # shows all duplicates
# deleting duplicates
data.drop_duplicates(inplace=True)
data.duplicated().sum() # shows no.of duplicates
```

```python
# checking and changing datatypes of columns
data.dtypes #shows datatypes
data["id"]=data["id"].astype(object)
data["host_id"]=data["host_id"].astype(object)
data.dtypes #shows datatypes
```

**5. Data Analysis**

*Univariate Analysis*

```python
# identfying tthe outliers in price
df = data[data["price"]<1500]
sns.boxplot(data=df,x="price")
plt.show()
```

```python
# price distribution
plt.figure(figsize=(8,4)) 
sns.histplot(data=df,x="price",kde=True,bins=100)
plt.title("price distribution")
plt.ylabel("Frequency")
plt.show()
```

```python
# availability distribution

plt.figure(figsize=(6,3))
sns.histplot(data=df,x="availability_365")
plt.title("Availability distribution")
plt.ylabel("Frequency")
plt.show()
```

```python
# average price per bed
df.groupby(by="neighbourhood_group")["price_per_bed"].mean()
```

**Feature Enginnering**

```python
# price per bed 
df["price_per_bed"]=df["price"]/df["beds"]
df.head()
```

**Bi variable Analysis :** *One variable depenency in another variable*

```python
# price dependency on neighbourhood
sns.barplot(data=df,x="neighbourhood_group",y="price",hue="room_type")
plt.show()
```

```python
# number of reviews and price relationship
plt.figure(figsize=(8,4))
plt.title("Locality reviews and price relationship")
sns.scatterplot(data=df,x="price",y="number_of_reviews",hue="neighbourhood_group")
plt.show()
```

```python
sns.pairplot(data=df,vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue='room_type')
```

```python
#Geographical Distribution of AirBnb Listing
plt.figure(figsize=(10,7))
sns.scatterplot(data=df,x="longitude",y="latitude",hue="room_type")
plt.title("Geographical Distribution of AirBnb Listing")
plt.show()
```

```python
# heat map - correlation of one variable with others for numerical column

corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr

plt.figure(figsize=(8, 4))
sns.heatmap(data=corr, annot=True)
plt.title("correlation of one variable with others for numerical column")
plt.show()
```
